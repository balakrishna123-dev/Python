# try – except


```python
# try – except
#  Syntax
# try:
#     # risky code
# except:
#     # handle error
# Example
try:
    print(10 / 0)
except:
    print("Error occurred")
```

```python
# try – except (Specific Exception)
#  Syntax
# try:
#     # risky code
# except ExceptionType:
#     # handle specific error

#  Example
try:
    print(10 / 0)
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
# Multiple except Blocks
# Syntax
# try:
# except ExceptionTy
#     # risky codepe1:
#     # handle error 1
# except ExceptionType2:
#     # handle error 2

# Example
try:
    x = int("abc")
    print(10 / x)
except ValueError:
    print("Invalid number")
except ZeroDivisionError:
    print("Division by zero")
```

```python
# try – except – else
#  Syntax
# try:
#     # risky code
# except ExceptionType:
#     # handle error
# else:
#     # runs if no error

#  Example
try:
    num = int("5")
except ValueError:
    print("Error")
else:
    print("Number is:", num)
```

```python
# try – except – finally
# Syntax
# try:
#     # risky code
# except ExceptionType:
#     # handle error
# finally:
#     # always runs

 #Example
try:
    print(10 / 2)
except ZeroDivisionError:
    print("Error")
finally:
    print("Execution completed")
```

```python
# Full Exception Handling (All Blocks)
#  Syntax
# try:
#     # risky code
# except ExceptionType1:
#     # handle error
# except ExceptionType2:
#     # handle error
# else:
#     # runs if no exception
# finally:
#     # always runs

#  Example
try:
    x = int(input("Enter number: "))
    print(10 / x)
except ValueError:
    print("Invalid input")
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("Calculation successful")
finally:
    print("Program ended")
```
