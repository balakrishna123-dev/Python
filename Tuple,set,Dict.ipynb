# Set

#A set in Python is a collection of unique elements that is unordered and unindexed.

```python
my_set = {10, 20, 20, 30}
print(my_set)     # {10, 20, 30}
```

2. Why Set? 

Sets are used when:

You want only unique items

You want to perform mathematical operations like union/intersection

```python
#How to Create a Set?

s = {1, 2, 3}
print(s)

#Using set()
c = set([1, 2, 3])
print(c)


# Empty set:

a= {}        # WRONG (this is dict)
b= set() 
print(a)
print(b)     # CORRECT
```

```python
A = {1, 2, 3}
B = {3, 4}

             # {1,2,3,4}
print(A.union(B))
print(A|B)
```

# Dictionary

what is Dictionary?

```python


# A dictionary in Python is a collection of key–value pairs.

# Keys → unique
# Values → can be duplicates
# Order → insertion ordered
# Mutable → can add, update, delete
```

Why we use Dictionary?


```python


# We use a dictionary when we want to store data in key → value format, like a real dictionary (word → meaning).

# Examples in real life:

# Student ID → Name

# Username → Password (facebook,student_credintial)

# Country → Capital

# Product ID → Product details
```

#How to Create a Dictionary?

```python
student = {"name": "Bala", "age": 25, "skills": "Python"}
print(student)
print(type(student))
```

```python
#Using dict()
student = dict(name="Bala", age=25)
print(student)
```

# Dictionary Comprehension

 #write a concise way for code in single using the same logic

```python
squares = {}
for x in range(5):
    squares[x] = x * x

print(squares)
```

#{ key_expression : value_expression  for item in iterable  if condition }

```python
squares = {x: x*x for x in range(5)}
print(squares)
```

```python
even_squares = {x: x*x for x in range(10) if x % 2 == 0}
print(even_squares)
```

# List Comprehension

```python
numbers = []
for x in range(5):
    numbers.append(x)
print(numbers)    
    
    
```

# Syntax:[ expression  for item in iterable  if condition ]


```python
numbers=[x for x in range(5)]
print(numbers)
```
