# if statement

```python
#Conditional Statement: Conditional Statement allows you to execute a block of code only when a specific condition is True.
if True:
    print("If Block")  
    print("Inside If")
```

```python
if False:
    print("if block")
    print("inside if")
```

# if-else statement

```python
 #If - Else Statement:  When the If - Else conditional statement is used, the Else block of code executes if the condition is False.

a = int(input()) # -1
if a > 0:  
    print("Positive")  
    
else:  
    print("Not Positive") 
```

```python
age = 14

if age >= 18:
    print("Adult")
else:
    print("Minor")
```

# if -elif-else statements

```python
# Elif Statement: Use the elif statement to have multiple conditional statements between if and else. The elif statement is optional.
 # syntax

if Condition A:
    block of code
elif Condition B:
    block of code
else:
    block of code
```

```python
score = 80

if score >= 90:
    print("Grade A")
elif score >= 70:
    print("Grade B")
else:
    print("Grade C")
```

# nested if( with in the if -condition you can use another if condition)

```python

age=75
if age >= 18:
    if age >= 60:
        print("Senior Citizen")
    else:
        print("Adult")
```

# range

Range: Generates a sequence of integers starting from 0. Stops before n (n is not included).

Syntax: range(n)



```python
for number in range(3):  
    print(number)
```

Range with Start and End: Generates a sequence of numbers starting from the start. Stops before the end (the end is not included).
Syntax: range(start, end)

```python
for number in range(5, 8):  
    print(number)
```

# loops

 Loops allow us to execute a block of code several times.


```python
 #1: for loop: for statement iterates over each item of a sequence.

Syntax:

for each_item in sequence:
    block of code
```

```python
nums = [1, 2, 3]

for n in nums:
    print(n)
```

```python
for i in range(5):
    print(i)
```

# while-loop( Allows us to execute a block of code several times as long as the condition is True)

```python
#While Loop: Allows us to execute a block of code several times as long as the condition is True.

a = 1
while a < 3:  
    a = a + 1  
    print(a)  
```

```python
count = 1
while count <= 5:
    print(count)
    count += 1
```

# loop Control Statements(break,pass,continue)

```python
#BREAK

#Stops the loop immediately.

for n in [1,2,3,4]:
    if n == 3:
        break
    print(n)
```

```python
# CONTINUE

#Skips current iteration.

for n in [1,2,3]:
    if n == 2:
        continue
    print(n)
```

```python
#PASS

#Does nothing (placeholder).

if n > 10:
    pass
```
