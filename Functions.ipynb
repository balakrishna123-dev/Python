What is a Function?

#Functions: Block of reusable code to perform a specific action.

#Defining a Function: Function is uniquely identified by the function_name.

```python
#def function_name():
 #   reusable code
```

#Calling a Function: The functional block of code is executed only when the function is called.

```python
# def function_name():
#     reusable code
# function_name()

# example:
def sum_of_two_number(a, b):
    print(a + b) # 5

sum_of_two_number(2, 3)
```

```python
#Function with Parameters (Input)

#Parameters = data you pass into a function.

def greet(name):
    print("Hello,", name)
greet("Bala")
```

# Function with Return Value 

```python


 # The function gives back a value using return.

def add(a, b):
    return a + b


result = add(5, 3)
print(result)
```
