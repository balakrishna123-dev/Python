# Naming Convention for Variables

```python

1. Must start with a letter or underscore _

✔ Valid:

name

_age

total1

✖ Invalid:

1name

@value

2. Cannot start with a number
1num = 10   # ❌ invalid
num1 = 10   # ✅ valid

3. Can contain only letters, numbers, and underscores

✔ first_name, marks_2024, total_amount
✖ first-name, total amount, price$

4. No spaces allowed

Use _ instead:

full_name = "Bala Krishna"  # ✔
full name = "Bala"          # ❌

**5. Variable names are case-sensitive
count = 10
Count = 20
COUNT = 30


All three are different variables.

6. Cannot use Python reserved keywords

Examples of keywords you cannot use:
for, while, if, class, def, return, try, True, False, None

Incorrect:

class = "Python"  # ❌

7. Use meaningful names (Best Practice)

✔ student_score
✔ total_price
✖ x, y, data123 (not meaningful)

8. Follow Snake Case (Python Convention)

Python uses snake_case:(in between we use underscore)

first_name = "Bala"
total_salary = 5000

9. Avoid using special characters

No: @, #, %, $, !, *, ., ?

10. Don’t use built-in function names

Avoid:

list = [1,2,3]    # ❌ overwrites built-in list()
sum = 10          # ❌ overwrites sum()
```

# bit_wise Operators

```python
#Bitwise operators work on bits (0s and 1s) of integers.

a = 5   # binary: 0101(
b = 3   # binary: 0011(

#Bitwise AND (&)

#Returns 1 if both bits are 1, else 0.

# 0101    #(5 )
# 0011    #(3)
# ----
# 0001 → 1

print(a & b)  # Output: 1

```

```python
# Bitwise OR (|)

# Returns 1 if any bit is 1, else 0.

# 0101
# 0011
# ----
# 0111 → 7

print(a | b)  # Output: 7
```

```python
# Bitwise XOR (^)

# Returns 1 if bits are different, else 0.

# 0101
# 0011
# ----
# 0110 → 6

print(a ^ b)  # Output: 6
```

```python
# NOT (~) (compliment)

# Flips all bits (0 → 1, 1 → 0) → also called 1’s complement.

# a = 5 → 0101
# ~a → 1010 → -6 (in Python, ~x = -(x+1))

print(~a)  # Output: -6
```
