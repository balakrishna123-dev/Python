# lambda Function

A lambda function is a small, anonymous (no name) function in Python that is written in one single line.

It is used when you need a simple function for a short time and don't want to formally define it using def.


#Syntax
lambda arguments: expression
lambda x:x**2


lambda → keyword

arguments → input parameters

expression → single line of logic (must return something)

```python
#Traditional function:
def add(x, y):
    return x + y
res=add(5,3)
print(res)
    
    
```

```python
#Same using lambda function:
add = lambda x, y: x + y
print(add(5, 3))   # Output: 8
```

Where Do We Use Lambda Functions? (Real use cases)

```python

#⃣ When you need a short, one-time function

#Example: Square of a number

square = lambda x: x * x
print(square(4))  # 16
```

# Map

```python


#Apply a function to each element in a list.

nums = [1, 2, 3, 4]
result = list(map(lambda x: x * 2, nums))
print(result)  # [2, 4, 6, 8]
```

# Filter

Keep only items that match a condition

Use: Select elements based on a condition (True/False).

Definition:
filter() keeps only the items for which the function returns True.

```python
With filter()

Filter elements based on a condition.

nums = [1, 2, 3, 4, 5]
even = list(filter(lambda x: x % 2 == 0, nums))
print(even)  # [2, 4]
```

# Reduce

reduce() — Combine all items into one value

Use: Perform cumulative operations to get one final result.
(You must import from functools)

```python
nums = [1, 2, 3, 4]

result = reduce(lambda x, y: x * y, nums)
print(result)
```

```python
 from functools import reduce
nums = [1, 2, 3, 4]

result = reduce(lambda x, y: x * y, nums)
print(result)   # 24
```
