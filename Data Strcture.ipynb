# String Slicing

```python
# String slicing means extracting a part (substring) from the string using this format:

#  string[start : end : step]

# start = index to begin

# end = index to stop (end index is NOT included)

# step = jump / skip characters
```

#Slice using start and end

```python

text = "PYTHON"
print(text[0:3])
```

#Slice without start (starts from 0)

```python

print(text[:4])
```

#Slice without end (goes till last)

```python

print(text[2:])
```

#Slice the entire string

```python

print(text[:])
```

#Step = 2 (skip characters)

```python

print(text[0:6:2])
```

# reverse a string

```python

print(text[::-1])
```

# List

What is a list?

#list  is  a collection of  items  which is orederd and changable(mutable) and allow duplicates in square brackets

# Methods In list

#Accessing List Items

```python
 #To access elements of a list, we use Indexing.

list_a = [5, "Six", 2, 8.2]
print(list_a[1]) # Six
```

#List Concatenation:

```python
#List Concatenation: Similar to strings, + operator concatenates lists.

list_a = [1, 2]
list_b = ["a", "b"]
list_c = list_a + list_b
print(list_c) # [1, 2, 'a', 'b']
```

#list Slicing

```python
#List Slicing: Obtaining a part of a list is called List Slicing.

list_a = [5, "Six", 2]
list_b = list_a[:2]
print(list_b) # [5, 'Six']
```

# Extend Slicing

```python
#Extended Slicing: Similar to string extended slicing, we can extract alternate items using the step.

list_a = ["R", "B", "G", "O", "W"]
list_b = list_a[0:5:3]
print(list_b) # ['R', 'O']
```

# Reverse a list

```python
 #Reversing a List: -1 for step will reverse the order of items in the list.

list_a = [5, 4, 3, 2, 1]
list_b = list_a[::-1]
print(list_b) # [1, 2, 3, 4, 5]
```

# Negative Slicing

```python
#Slicing With Negative Index: You can also specify negative indices while slicing a List.

list_a = [5, 4, 3, 2, 1]
list_b = list_a[-3:-1]
print(list_b) # [3, 2]
```

# Changing List Items(because of Mutability)

```python
thislist = ["apple", "banana", "cherry"]
thislist[1] = "blackberrry"
print(thislist)
```

# add list items;

#append:( add a single value to the list at the end of the  list)

```python

#Adds one element at the end.

nums = [1, 2, 3]
nums.append(4)
print(nums)
# [1, 2, 3, 4]
```

```python
# Extend : add multiple value to a list
```

```python
nums.extend([5,6])
print(nums)
```

```python
#insert: Adds element at a specific index.

nums.insert(1, 10)
print(nums)
# [1, 10, 2, 3]
```

# Remove elements

```python
#  Deletes first occurrence of a value.
num=[1, 10, 2, 3, 4, 5, 6,10]
num.remove(10)
print(num)
```

```python

#pop():Removes & returns element at index (default = last).

print(num.pop())
print(num.pop(0))
```

```python
#clear()

#Deletes all elements.

num.clear()
print(num)
```
