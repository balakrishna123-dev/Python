What is sqlite3?

sqlite3 is a built-in Python module

It allows Python to connect to an SQLite database

```python
import sqlite3
```

```python
conn = sqlite3.connect("student.db")
print("Database connected successfully")
```

```python
# Create a Cursor Object
# Cursor is used to execute SQL commands
# What is a Cursor?

# Cursor is like a pointer or messenger

# It sends SQL commands to the database

# It also receives results from the database

cursor = conn.cursor()
print(cursor)
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    marks REAL
)
""")
```

```python
students = [
    ('Anil', 22, 88.0),
    ('Sunil', 20, 90.5),
    ('Ravi', 23, 76.0)
]

cursor.executemany("""
INSERT INTO students (name, age, marks)
VALUES (?, ?, ?)
""", students)
print(students)
```

```python
cursor.execute("SELECT * FROM students")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
cursor.execute("SELECT * FROM students WHERE age > 21")
row = cursor.fetchone()
print(row)
```

```python
cursor.execute("""
DELETE FROM students
WHERE name = 'Ravi'
""")

conn.commit()
```
